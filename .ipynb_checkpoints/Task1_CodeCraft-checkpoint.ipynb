```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("test.csv")
```

```python
data
```

```python
data['Sex'].value_counts().plot(kind='bar', title='Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
data['Age'].dropna().plot(
    kind='hist', 
    bins=15, 
    rwidth=0.9, 
    color='orange',
    edgecolor='black',
    title='Age Distribution'
)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
data['Pclass'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, shadow=True)
plt.title('Passenger Class Distribution')
plt.ylabel('')  # Remove default ylabel
plt.tight_layout()
plt.show()
```
